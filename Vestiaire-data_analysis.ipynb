```python
import kagglehub
import pandas as pd
import os

path = kagglehub.dataset_download("justinpakzad/vestiaire-fashion-dataset")
print("Path to dataset files:", path)
df = pd.read_csv(os.path.join(path, 'vestiaire.csv'), low_memory=False)
```

```python
df['item_type'] = df['product_type'].str.strip().str.split().str[-1].str.lower()
```

```python
## Price Summary by Product Condition

condition_summary = df.groupby('product_condition')[['price_usd', 'seller_price', 'seller_earning']].agg(['count', 'mean', 'median', 'min', 'max'])
print("Condition-Based Price Summary:")
print(condition_summary)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Prepare data
avg_price = df.groupby('product_condition')['price_usd'].mean().sort_values(ascending=False)
conditions = avg_price.index
values = avg_price.values

# Set Seaborn theme
sns.set(style="whitegrid")

# Create plot
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=values, y=conditions, palette="viridis")

# Add value labels
for i, v in enumerate(values):
    ax.text(v + 5, i, f"${v:,.0f}", color='black', va='center', fontweight='bold')

# Styling
plt.title("💼 Average Resale Price by Product Condition", fontsize=16, weight='bold')
plt.xlabel("Average Price (USD)", fontsize=12)
plt.ylabel("Product Condition", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
## Price Summary by Product Category

category_summary = df.groupby('product_category')[['price_usd', 'seller_price', 'seller_earning']].agg(['count', 'mean', 'median', 'min', 'max'])\
                     .sort_values(('price_usd', 'mean'), ascending=False)
print("\nCategory-Based Price Summary:")
print(category_summary)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Prepare data
avg_price_by_category = df.groupby('product_category')['price_usd'].mean().sort_values(ascending=False)

# Filter categories with at least 5 listings (optional, for cleaner plot)
category_counts = df['product_category'].value_counts()
valid_categories = category_counts[category_counts >= 5].index
avg_price_by_category = avg_price_by_category[avg_price_by_category.index.isin(valid_categories)]

# Set visual style
sns.set(style="whitegrid")

# Plot
plt.figure(figsize=(12, 8))
ax = sns.barplot(x=avg_price_by_category.values, y=avg_price_by_category.index, palette="magma")

# Add value labels
for i, v in enumerate(avg_price_by_category.values):
    ax.text(v + 5, i, f"${v:,.0f}", color='black', va='center', fontweight='bold')

# Title and labels
plt.title("👜 Average Resale Price by Product Category", fontsize=16, weight='bold')
plt.xlabel("Average Price (USD)", fontsize=12)
plt.ylabel("Product Category", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
## Price Summary by Brand

brand_summary = df.groupby('brand_name')[['price_usd', 'seller_price', 'seller_earning']].mean()
brand_summary['avg_margin'] = brand_summary['price_usd'] - brand_summary['seller_earning']
brand_summary = brand_summary.sort_values('avg_margin', ascending=False)
print("\nBrand-Level Price & Margin Summary:")
print(brand_summary)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter brands with at least 5 listings
brand_counts = df['brand_name'].value_counts()
valid_brands = brand_counts[brand_counts >= 5].index
df_valid_brands = df[df['brand_name'].isin(valid_brands)]

# Compute average resale price by brand
brand_price = df_valid_brands.groupby('brand_name')['price_usd'].mean().sort_values(ascending=False).head(15)

# Plot
plt.figure(figsize=(12, 7))
sns.set(style="whitegrid")
ax = sns.barplot(x=brand_price.values, y=brand_price.index, palette="crest")

# Add value labels
for i, v in enumerate(brand_price.values):
    ax.text(v + 5, i, f"${v:,.0f}", color='black', va='center', fontweight='bold')

# Style
plt.title("👛 Top 15 Brands by Average Resale Price", fontsize=16, weight='bold')
plt.xlabel("Average Resale Price (USD)", fontsize=12)
plt.ylabel("Brand", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
## Price Distribution by Seller Country

country_summary = df.groupby('seller_country')[['price_usd', 'seller_price', 'seller_earning']].agg(['count', 'mean', 'median'])\
                    .sort_values(('price_usd', 'mean'), ascending=False)
print("\nSeller Country-Based Price Summary:")
print(country_summary)
```

```python
## Outlier Brands

# Only consider brands with at least 5 listings to avoid noise
brand_counts = df['brand_name'].value_counts()
valid_brands = brand_counts[brand_counts >= 5].index
brand_price_stats = df[df['brand_name'].isin(valid_brands)].groupby('brand_name')['price_usd'].mean()

# Sort to identify high and low outliers
top_brands = brand_price_stats.sort_values(ascending=False).head(10)
bottom_brands = brand_price_stats.sort_values().head(10)

print("Top 10 Most Expensive Brands (Average Resale Price):")
print(top_brands)

print("\nBottom 10 Least Expensive Brands (Average Resale Price):")
print(bottom_brands)
```

```python
## Outlier Categories

# Only consider categories with at least 5 listings
category_counts = df['product_category'].value_counts()
valid_cats = category_counts[category_counts >= 5].index
cat_price_stats = df[df['product_category'].isin(valid_cats)].groupby('product_category')['price_usd'].mean()

# Sort to identify outliers
top_categories = cat_price_stats.sort_values(ascending=False).head(10)
bottom_categories = cat_price_stats.sort_values().head(10)

print("Top 10 Most Expensive Categories (Average Resale Price):")
print(top_categories)

print("\nBottom 10 Least Expensive Categories (Average Resale Price):")
print(bottom_categories)
```

```python
## Price Summary by Item Type

# Only include item types with at least 5 listings
item_counts = df['item_type'].value_counts()
valid_items = item_counts[item_counts >= 5].index
item_price_stats = df[df['item_type'].isin(valid_items)].groupby('item_type')['price_usd'].mean()

# Top and bottom 10 item types by average price
top_items = item_price_stats.sort_values(ascending=False).head(10)
bottom_items = item_price_stats.sort_values().head(10)

print("Top 10 Most Expensive Item Types (Average Resale Price):")
print(top_items)

print("\nBottom 10 Least Expensive Item Types (Average Resale Price):")
print(bottom_items)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Prepare data for top 10 item types
top_items = item_price_stats.sort_values(ascending=False).head(10)

# Plot
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
ax = sns.barplot(x=top_items.values, y=top_items.index, palette="rocket")

# Add value labels
for i, v in enumerate(top_items.values):
    ax.text(v + 5, i, f"${v:,.0f}", color='black', va='center', fontweight='bold')

# Style
plt.title("🧳 Top 10 Item Types by Average Resale Price", fontsize=16, weight='bold')
plt.xlabel("Average Resale Price (USD)", fontsize=12)
plt.ylabel("Item Type", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
##  Brand-Level Summary

# Filter brands with at least 5 products
brand_counts = df['brand_name'].value_counts()
valid_brands = brand_counts[brand_counts >= 5].index
df_valid_brands = df[df['brand_name'].isin(valid_brands)]

# Aggregate brand-level metrics
brand_summary = df_valid_brands.groupby('brand_name').agg(
    num_listings=('product_id', 'count'),
    avg_price_usd=('price_usd', 'mean'),
    avg_seller_price=('seller_price', 'mean'),
    avg_earning=('seller_earning', 'mean'),
    total_sold=('sold', 'sum')
)

# Add margin and conversion rate
brand_summary['avg_margin'] = brand_summary['avg_price_usd'] - brand_summary['avg_earning']
brand_summary['conversion_rate'] = brand_summary['total_sold'] / brand_summary['num_listings']

# Sort by average price
brand_summary = brand_summary.sort_values('avg_price_usd', ascending=False)

print("Top Brands by Average Resale Price:")
print(brand_summary.head(10))
```

```python
## Brands with Highest Conversion Rate

top_conversion = brand_summary.sort_values('conversion_rate', ascending=False).head(10)
print("Top 10 Brands by Conversion Rate:")
print(top_conversion[['conversion_rate', 'num_listings']])
```

```python
## Distribution of Product Conditions

# Count of listings per condition
condition_counts = df['product_condition'].value_counts()
print("Listing Count by Product Condition:")
print(condition_counts)
```

```python
## Price & Sales by Condition

condition_summary = df.groupby('product_condition').agg(
    num_listings=('product_id', 'count'),
    avg_price_usd=('price_usd', 'mean'),
    avg_likes=('product_like_count', 'mean'),
    sold_count=('sold', 'sum'),
    conversion_rate=('sold', 'mean')  # same as sold_count / num_listings
).sort_values('avg_price_usd', ascending=False)

print("\nCondition-Based Summary:")
print(condition_summary)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Prepare data
avg_price_by_condition = condition_summary['avg_price_usd'].sort_values(ascending=False)

# Plot
plt.figure(figsize=(9, 5))
sns.set(style="whitegrid")
ax = sns.barplot(x=avg_price_by_condition.values, y=avg_price_by_condition.index, palette="light:#5A9")

# Add value labels
for i, v in enumerate(avg_price_by_condition.values):
    ax.text(v + 5, i, f"${v:,.0f}", color='black', va='center', fontweight='bold')

# Style
plt.title("🔧 Average Resale Price by Product Condition", fontsize=16, weight='bold')
plt.xlabel("Average Resale Price (USD)", fontsize=12)
plt.ylabel("Product Condition", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sort by conversion rate for cleaner look
sorted_conversion = condition_summary['conversion_rate'].sort_values(ascending=False)

# Set visual style
sns.set(style="whitegrid")

# Plot
plt.figure(figsize=(9, 5))
ax = sns.barplot(
    x=sorted_conversion.values,
    y=sorted_conversion.index,
    palette="ch:s=.25,rot=-.25"
)

# Add value labels
for i, v in enumerate(sorted_conversion.values):
    ax.text(v + 0.01, i, f"{v:.1%}", color='black', va='center', fontweight='bold')

# Style
plt.title("📈 Conversion Rate by Product Condition", fontsize=16, weight='bold')
plt.xlabel("Conversion Rate", fontsize=12)
plt.ylabel("Product Condition", fontsize=12)
plt.xlim(0, 1)  # Conversion is a proportion (0–1)
plt.tight_layout()
plt.show()
```
